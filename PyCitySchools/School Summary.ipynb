```python
import pandas as pd
import numpy as np
```

```python
schoolfile = "C:/Users/clara/Documents/GitHub/pandas-challenge/Starter_Code/PyCitySchools/Resources/schools_complete.csv"
```

```python
schoolfile_df = pd.read_csv(schoolfile)

#change column name before it becomes the index
schoolfile_df = schoolfile_df.rename(columns={"school_name": "School name"})

#change index of school file to allow for concatenation at the end
schoolind_df = schoolfile_df.set_index("School name")

#delete school id column
del schoolind_df["School ID"]

#add budget per student column
budperst = schoolind_df["budget"] / schoolind_df["size"]
schoolind_df["Budget per student in $"] = budperst

#rename columns
schoolind_df = schoolind_df.rename(columns={"type": "School type"})
schoolind_df = schoolind_df.rename(columns={"size": "Total students"})
schoolind_df = schoolind_df.rename(columns={"budget": "Total school budget in $"})

schoolind_df = schoolind_df.sort_values(by=['Budget per student in $'], ascending=True)

schoolind_df.head(15)
```

```python
studentfile = "C:/Users/clara/Documents/GitHub/pandas-challenge/Starter_Code/PyCitySchools/Resources/students_complete.csv"
```

```python
studentfile_df = pd.read_csv(studentfile)
studentfile_df.head()
```

```python
#series for math mean
grmathmean_df = studentfile_df.groupby(["school_name"]).mean()["math_score"].round(1)

#series for read mean
grreadmean_df = studentfile_df.groupby(["school_name"]).mean()["reading_score"].round(1)

#series for % pass math
grpassmath_df = studentfile_df[studentfile_df["math_score"] >= 70].groupby(["school_name"]).count()["math_score"]
grpassmath_df = round(grpassmath_df / schoolind_df["Total students"]*100, 1)

#series for % pass read
grpassread_df = studentfile_df[studentfile_df["reading_score"] >= 70].groupby(["school_name"]).count()["reading_score"]
grpassread_df = round(grpassread_df / schoolind_df["Total students"]*100, 1)

#series for % pass read & math
grpassreadandmath_df = studentfile_df[(studentfile_df['math_score'] >= 70) & (studentfile_df['reading_score'] >= 70)].groupby(['school_name']).count()["student_name"]
grpassreadandmath_df = round(grpassreadandmath_df / schoolind_df["Total students"]*100, 1)
```

```python
#concatenate dataframes to put together the school summary data frame
per_school_summary = pd.concat([schoolind_df, grmathmean_df, grreadmean_df, grpassmath_df, grpassread_df, grpassreadandmath_df], 1)

#rename columns
#schoolind_df = schoolind_df.rename(columns={"type": "School type"})
per_school_summary = per_school_summary.rename(columns={"math_score": "Average math score"})
per_school_summary = per_school_summary.rename(columns={"reading_score": "Average reading score"})
per_school_summary = per_school_summary.rename(columns={0: "% passing math"})
per_school_summary = per_school_summary.rename(columns={1: "% passing read"})
per_school_summary = per_school_summary.rename(columns={2: "% overall passing"})
per_school_summary = per_school_summary.sort_values(by=['Budget per student in $'], ascending=True)
per_school_summary
```

```python
#Highest-Performing Schools (by % Overall Passing)
per_school_summary = per_school_summary.sort_values(by=['% overall passing'], ascending=False)
top_schools = per_school_summary.head(5)
top_schools
```

```python
#Lowest-Performing Schools (by % Overall Passing)
per_school_summary = per_school_summary.sort_values(by=['% overall passing'])
bottom_schools = per_school_summary.head(5)
bottom_schools
```

```python
#Reading Scores by Grade
reading_scores_by_grade = studentfile_df.groupby(["school_name", "grade"]).mean()["reading_score"].round(1)
reading_scores_by_grade.unstack(level=-1)
```

```python
#Scores by School Spending
#defining bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_school_summary["Budget per student in $"],spending_bins,labels=labels)
per_school_summary = per_school_summary.sort_values(by=['Spending Ranges (Per Student)'], ascending=False)
per_school_summary
```

```python
#use code given in exercise
school_spending_df = per_school_summary
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average math score"].round(2)
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average reading score"].round(2)
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% passing math"].round(2)
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% passing read"].round(2)
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% overall passing"].round(2)
#create dataframe
spending_summary = pd.concat([spending_math_scores,spending_reading_scores,spending_passing_math,spending_passing_reading,overall_passing_spending],1)
spending_summary
```

```python
#Scores by School Size (5 points)
#defining bins and labels
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
per_school_summary["Size label"] = pd.cut(per_school_summary["Total students"],size_bins,labels=labels)
per_school_summary = per_school_summary.sort_values(by=['Size label'], ascending=False)
per_school_summary
```

```python
#Scores by School Type (5 points)
#mathbygrade = studentfile_df.groupby(["school_name", pd.Grouper( key="grade")]).mean()["math_score"].round(1)

schooltype_math_scores = school_spending_df.groupby(["School type"]).mean()["Average math score"].round(2)
schooltype_reading_scores = school_spending_df.groupby(["School type"]).mean()["Average reading score"].round(2)
schooltype_passing_math = school_spending_df.groupby(["School type"]).mean()["% passing math"].round(2)
schooltype_passing_read = school_spending_df.groupby(["School type"]).mean()["% passing read"].round(2)
schooltype_passing_spending = school_spending_df.groupby(["School type"]).mean()["% overall passing"].round(2)

#create dataframe
type_summary = pd.concat([schooltype_math_scores,schooltype_reading_scores,schooltype_passing_math,schooltype_passing_read,schooltype_passing_spending],1)
type_summary
```
